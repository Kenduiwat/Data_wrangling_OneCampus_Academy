```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
file_path = r'C:\Users\vkori\Desktop\OneCampus\Muskets_data.csv'

# Read the CSV file with low_memory=False
df = pd.read_csv(file_path, low_memory=False)
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.head()
```

```python
# Extract player names from the "playerurl" column and create a new column "Player Name"
df['Player Name'] = df['playerurl'].str.extract(r'/([^/]+)/\d+/$')

# Display the DataFrame with the new "Player Name" column
print(df[['Player Name']])
```

```python
df.head()
```

```python
# Create a new column "Player Status" based on the "contract" column
df['Player Status'] = df['contract'].apply(lambda x: 'Active' if '202' in x else ('Free' if '202' not in x else 'On Loan'))

# Display the DataFrame with the new "Player Status" column
print(df[['Player Status']])
```

```python
df.head()
```

```python
# Check for rows where "Player Status" is "On Loan"
on_loan_players = df[df['Player Status'] == 'On Loan']

# Display the rows where "Player Status" is "On Loan"
print(on_loan_players)
```

```python
# Check for rows where "Player Status" is "Free"
free_players = df[df['Player Status'] == 'Free']

# Display the rows where "Player Status" is "Free"
print(free_players)

```

```python
# Split the positions in the "position" column and create dummy columns
positions_df = df['positions'].str.get_dummies(sep=', ')

# Concatenate the new columns with the original DataFrame
df = pd.concat([df, positions_df], axis=1)

# Display the resulting DataFrame
print(df)
```

```python
df.head()
```

```python
print(df)
```

```python
# Convert column names to lowercase
df.columns = df.columns.str.lower()

# Convert the "value," "wage," and "release clause" columns to string
df['value'] = df['value'].astype(str)
df['wage'] = df['wage'].astype(str)
df['release clause'] = df['release clause'].astype(str)

# Convert the "value" column to float
df['value'] = df['value'].str.extract(r'€([\d.]+)([KM]+)').apply(lambda x: pd.to_numeric(x[0]) * 1e3 if x[1] == 'M' else pd.to_numeric(x[0]), axis=1).astype(float)

# Convert the "wage" column to float
df['wage'] = df['wage'].str.extract(r'€([\d.]+)([KM]+)').apply(lambda x: pd.to_numeric(x[0]) * 1e3 if x[1] == 'M' else pd.to_numeric(x[0]), axis=1).astype(float)

# Convert the "release clause" column to float
df['release clause'] = df['release clause'].str.extract(r'€([\d.]+)([KM]+)').apply(lambda x: pd.to_numeric(x[0]) * 1e3 if x[1] == 'M' else pd.to_numeric(x[0]), axis=1).astype(float)
```

```python
df.head()
```

```python
# Convert the "hits" column to lowercase, then replace 'k' with 'e3' and convert to float
df['hits'] = df['hits'].astype(str).str.lower().str.replace('k', 'e3', regex=True).astype(float)

# Display the data types of the columns
print(df.dtypes)
```

```python
df.head()
```

```python
# Convert columns to lowercase (assuming they're already in the DataFrame)
df.columns = df.columns.str.lower()

# Convert the columns to numeric
df['height'] = pd.to_numeric(df['height'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['wage'] = pd.to_numeric(df['wage'], errors='coerce')
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['release clause'] = pd.to_numeric(df['release clause'], errors='coerce')

# Create labels for height intervals (every 10 units)
height_labels = [f"{i}-{i+9}" for i in range(140, 220, 10)]
df['height_category'] = pd.cut(df['height'], bins=range(140, 230, 10), labels=height_labels)

# Create labels for weight intervals (every 10 kg)
weight_labels = [f"{i}-{i+9}" for i in range(40, 120, 10)]
df['weight_category'] = pd.cut(df['weight'], bins=range(40, 130, 10), labels=weight_labels)

# Create labels for wage intervals (every 50k)
wage_labels = [f"{i}-{i+49}k" for i in range(0, 350, 50)]
df['wage_category'] = pd.cut(df['wage'], bins=range(0, 400, 50), labels=wage_labels)

# Create labels for value intervals (every 50m)
value_labels = [f"{i}-{i+49}m" for i in range(0, 300, 50)]
df['value_category'] = pd.cut(df['value'], bins=range(0, 350, 50), labels=value_labels)

# Create labels for release clause intervals (every 50m)
release_labels = [f"{i}-{i+49}m" for i in range(0, 300, 50)]
df['release_clause_category'] = pd.cut(df['release clause'], bins=range(0, 350, 50), labels=release_labels)
```
